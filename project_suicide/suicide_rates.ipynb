```python
import pandas as pd
from sqlalchemy import create_engine
```

```python

suicide_data = "./resources/master.csv"

suicide_data_df = pd.read_csv(suicide_data)
suicide_data_df
```

```python
suicide_data_c_df = suicide_data_df.loc[suicide_data_df["year"] >=2010]
suicide_data_c_df
```

```python
suicide_data_c_df.columns

suicide_data_clean_df = suicide_data_c_df[['country', 'year', 'sex',
       'suicides/100k pop',
       'gdp_per_capita ($)']]
suicide_data_clean_df
```

```python
suicide_data_r_df = suicide_data_clean_df.rename(columns = {'suicides/100k pop':'suicide_per_100k_popn'})
suicide_data_r_df
```

```python
suicide_data_renamed_df = suicide_data_r_df.rename(columns = {'gdp_per_capita ($)':'gdp_per_capita_usd'})
suicide_data_renamed_df
```
