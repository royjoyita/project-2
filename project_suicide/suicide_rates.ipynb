```python
create table covid_usa(
date date primary key,
total_cases int,
new_cases int,
total_deaths int,
new_deaths int
)

create table vaccination_usa(
date date primary key,
vaccine varchar,
total_vaccinations int
)

create table us_vaccination_no(
date date primary key,
total_distributed int,
people_fully_vaccinated int
)

SELECT covid_usa.date,covid_usa.total_cases, covid_usa.new_cases, covid_usa.total_deaths, covid_usa.new_deaths,
us_vaccination_no.total_distributed, us_vaccination_no.people_fully_vaccinated, vaccination_usa.vaccine, vaccination_usa.total_vaccinations
FROM covid_usa
INNER JOIN us_vaccination_no ON covid_usa.date = us_vaccination_no.date
INNER JOIN vaccination_usa ON covid_usa.date = vaccination_usa.date
WHERE covid_usa.date > '2021-01-01' and covid_usa.date < '2021-03-01'
ORDER BY date ASC
;
```
